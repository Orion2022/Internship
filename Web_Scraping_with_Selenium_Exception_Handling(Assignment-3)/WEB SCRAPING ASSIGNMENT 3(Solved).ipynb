# 1. Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user.

```python
#path for chromedriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')

# to open the homepage of Amazon.in
url = 'https://www.amazon.in'
driver.get(url)

time.sleep(3)

#user input
user_inp=str(input("Enter the product name you want to search: "))

#finding the search option on our web driver
search=driver.find_element_by_id("twotabsearchtextbox")
search.send_keys(user_inp)

# clicking the search button
search_btn=driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']/span/input")
search_btn.click()
```

# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# scraping all the product urls
urls =[]
for i in range(0,3): # upto 3 pages
    for x in driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']"):
        urls.append(x.get_attribute('href'))
    driver.find_element_by_xpath("//li[@class='a-normal'][1]/a").click()
    time.sleep(3)
```

```python
#Making empty lists
brand=[]
prod_name=[]
Rating=[]
rating_num=[]
Price= []
Return=[]
exp_deliv = []
availability=[]
prod_url=[]
Other_Details = []

#extracting all the details one after another
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    # getting the brand name
    try:
        brand_name=driver.find_element_by_xpath("//td[@class='a-span9']/span")
        brand.append(brand_name.text)
    except NoSuchElementException:
        brand.append('-')    
                
      # getting the product name
    try:
        product=driver.find_element_by_id("productTitle")
        prod_name.append(product.text)
    except NoSuchElementException:
        prod_name.append('-')         
        
      # getting the rating
    try:
        rate=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")
        Rating.append(rate.text)
    except NoSuchElementException:
        Rating.append("No Rating")       
                
      # getting the rating number
    try:
        rate_num=driver.find_element_by_id("acrCustomerReviewText")
        rating_num.append(rate_num.text)
    except NoSuchElementException:
        rating_num.append('-')         
                
      # getting the price
    try:
        prod_price=driver.find_element_by_id("priceblock_ourprice")
        Price.append(prod_price.text)
    except NoSuchElementException:
        Price.append('-')      
                
     # getting the return/exchange
    try:
        ret_exc=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        Return.append(ret_exc.get_attribute("href"))
    except NoSuchElementException:
        Return.append('-') 
        
     # getting the expected delivery
    try:
        ex_de=driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b")
        exp_deliv.append(ex_de.text)
    except NoSuchElementException:
        exp_deliv.append('-')           
               
     # getting the availability
    try:
        avail=driver.find_element_by_id("availability")
        availability.append(avail.text)
    except NoSuchElementException:
        availability.append('-')       
                           
     # getting the product url
    try:
        pro_url=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']/a")
        prod_url.append(pro_url.get_attribute("href"))
    except NoSuchElementException:
        prod_url.append('-')
        
     # getting the other details
    try:
        othr_det=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']/li/span")
        Other_Details.append(othr_det.text)
    except NoSuchElementException:
        Other_Details.append('-')
        
        
# Make dataframe
df = pd.DataFrame({})
df['Brand Name'] = brand
df['Product Name'] = prod_name
df['Rating'] = Rating
df['Rating Number'] = rating_num
df['Price'] = Price
df['Return/Exchange'] = Return
df['Expected Delivery'] = exp_deliv
df['Availability'] = availability
df['Product URLs'] = prod_url
df['Other Details'] = Other_Details     
df.head(5)
```

# 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
#path for chromedriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')

# to open flipkart webpage
url = "https://www.flipkart.com/"
driver.get(url)
time.sleep(2)

#closing login popup
try:
    login_btn = driver.find_element_by_xpath('//div[@class="_2QfC02"]//button')
    login_btn.click()
except NoSuchElementException : 
    print("No Login Page")
    
#initializing search bar
search_bar = driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input")
search_bar.send_keys("Realme")

#initializing search button
search_btn = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
time.sleep(3)
```

```python
#Creating empty lists to store the data
product_urls=[]
brand_name=[]
phone_name=[]
phone_color=[]
RAM=[]
ROM=[]
prim_cam=[]
sec_cam=[]
dis_size=[]
dis_resolution=[]
processor=[]
processor_core=[]
battery=[]
price=[]

# fetching urls of smartphone
urls=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in urls:
    product_urls.append(i.get_attribute('href'))
for i in product_urls:
    driver.get(i)
    
    #expanding the specifications
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    
    #scraping brand
    try:
        brand=driver.find_element_by_xpath("//span[@class='B_NuCI']").text.split()
        brand_name.append(brand[0])
    except NoSuchElementException:
        brand_name.append('-')
        
    #scraping the phone name
    try:
        phone=driver.find_element_by_xpath("//h1[@class='yhB1nd']//span")
        phone_name.append(phone.text)
    except NoSuchElementException:
        phone_name.append('-')
    
    #scraping the color
    try:
        color=driver.find_element_by_xpath("//div[@class='_1UhVsV']/div[1]/table/tbody/tr[4]/td[2]")
        phone_color.append(color.text)
    except NoSuchElementException:
        phone_color.append('-')
    
    #scraping RAM
    try:
        RAM.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[2]/td[2]/ul/li').text)
    except NoSuchElementException:
        RAM.append('-')
    
    #scraping Storage(ROM)
    try:
        ROM.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']/div[4]/table/tbody/tr[1]/td[2]").text)
    except NoSuchElementException:
        ROM.append('-')
    
    #scraping Primary camera
    try:
        prim_cam.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']/div[5]/table/tbody/tr[2]/td[2]").text)
    except NoSuchElementException:
        prim_cam.append('-')
    
    #scraping secondary camera
    try:
        sec_cam.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']/div[5]/table/tbody/tr[5]/td[2]").text)
    except NoSuchElementException:
        sec_cam.append('-')
    
    #scraping display size
    try:
        dis_size.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']/div[2]/table/tbody/tr[1]/td[2]").text)
    except NoSuchElementException:
        dis_size.append('-')
    
    #scraping display resolution
    try:
        dis_resolution.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']/div[2]/table/tbody/tr[2]/td[2]").text)
    except NoSuchElementException:
        dis_resolution.append('-')
    
    #scraping processor
    try:
        processor.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']/div[3]/table/tbody/tr[2]/td[2]").text)
    except NoSuchElementException:
        processor.append('-')
    
    #scraping processor core
    try:
        processor_core.append(driver.find_element_by_xpath('//div[@class="_3k-BhJ"][3]/table[1]/tbody/tr[2]/td[1]').text)
    except NoSuchElementException:
        processor_core.append('-')
    
    #scraping battery capacity
    try:
        battery.append(driver.find_element_by_xpath("//div[@class='_3k-BhJ'][10]/div").text)
    except NoSuchElementException:
        battery.append('-')
    
    #scraping the price of the phone
    try:
        price.append(driver.find_element_by_xpath("//div[@class='dyC4hf']/div/div/div").text)
    except NoSuchElementException:
        price.append('-')
```

```python
#Making dataframe
flip=pd.DataFrame({})
flip['Brand']=brand_name
flip['Smartphone']=phone_name
flip['Price']=price
flip['Color']=phone_color
flip['RAM']=RAM
flip['ROM']=ROM
flip['Primary camera']=prim_cam
flip['Secondary camera']=sec_cam
flip['Battery']=battery
flip['Display size']=dis_size
flip['Display resolution']=dis_resolution
flip['Processor']=processor
flip['Processor core']=processor_core
flip['Product_url']=product_urls
flip
```

```python
# Exporting the above dataframe into csv-format 
flip.to_csv(r"C:\Users\HP\Flip_robo_assignments\Smartphone_Details_From_Flipkart.csv", index=False)
```

```python
# empty lists to scraping data
date=[]
name=[]
industry=[]
sub_ver=[]
loc=[]
inv=[]
inv_type=[]
amount=[]

for j in range (48,51):
    dropdown=driver.find_element_by_xpath(f"//div[@class='dataTables_length']//select[@name='tablepress-{j}_length']")
    dropdown.send_keys("100")
    

    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-2']"): #date
        date.append(i.text)
    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-3']"): #startup name
        name.append(i.text)
    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-4']"): #industry name
        industry.append(i.text)
    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-5']"): #Sub-vertical
        sub_ver.append(i.text)
    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-6']"): #location
        loc.append(i.text)
    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-7']"): #investor
        inv.append(i.text)
    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-8']"): # investment type
        inv_type.append(i.text)
    for i in driver.find_elements_by_xpath(f"//table[@id='tablepress-{j}']//td[@class='column-9']"): # amount
        amount.append(i.text)
        
# Making dataframe from all the collected information
df = pd.DataFrame({"Date":date,
                 "Startup-name":name,
                 "Industry/Vertical":industry,
                 "Sub-vertical":sub_ver,
                 "Location":loc,
                 "Investor":inv,
                 "Investment Type":inv_type,
                 "Amount(in USD)":amount})
df
    
```

# 10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
#path for chromedriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')

# to open the webpage
url = "https://www.hostelworld.com/"
driver.get(url)
time.sleep(2)
```

```python
#finding the location search bar
search_bar = driver.find_element_by_xpath("//input[@id='location-text-input-field']")
search_bar.send_keys('London')
```

```python
#selecting london
London = driver.find_element_by_xpath('/html/body/div[1]/div/div/div[1]/div[1]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div')
London.click()
time.sleep(1)
```

```python
# clicking on search button
srch_button = driver.find_element_by_id('search-button')
srch_button.click()
```

```python
# empty lists to store the data
Hostel_name = []
Distance = []
Pvt_prices = []
Dorms_price = []
Rating = []
Reviews = []
Over_all = []
Facilities = []
Description =[]
Product_url = []
```

```python
for i in driver.find_elements_by_xpath("//div[@class = 'pagination-item pagination-next']"):
    i.click()
    time.sleep(2)
    
    # scraping hostel name
    try:
        name = driver.find_elements_by_xpath("//h2[@class='title title-6']")
        for i in name:
            Hostel_name.append(i.text)
    except NoSuchElementException:
        Hostel_name.append('-')
        
    time.sleep(2)
        
    # scraping the distance from city centre    
    try:
        dist = driver.find_elements_by_xpath("//span[@class='description']")
        for i in dist:
            Distance.append(i.text.replace('Hostel - ',''))
    except NoSuchElementException:
        Distance.append('-')
     
    time.sleep(2)   
        
    # scraping the privates from price
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            pvt_price = driver.find_element_by_xpath("//a[@class='prices']//div[1]//div")
            Pvt_prices.append(pvt_price.text)
        except NoSuchElementException:
            Pvt_prices.append('-')
    
    time.sleep(2)
    
    # scraping the dorms from price
    for i in driver.find_elements_by_xpath("//div[@class='prices-col']"):
        try:
            dorms = driver.find_element_by_xpath("//a[@class='prices']//div[2]//div")
            Dorms_price.append(dorms.text)
        except NoSuchElementException:
            Dorms_price.append('-')
    
    time.sleep(2)        
    
    # scraping the facilities
    try:
        faci = driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
        for i in faci:
            Facilities.append(i.text)
    except NoSuchElementException:
        Facilities.append('-')
    
    time.sleep(2)
    
    # scraping the urls of each hostel
    pro_url = driver.find_elements_by_xpath("//div[@class='prices-col']//a[2]")
    for i in pro_url:
        Product_url.append(i.get_attribute('href'))
    
    time.sleep(2)

for i in Product_url:
    driver.get(i)
    time.sleep(2)
    
    # clicking on the show more button for description
    try:
        driver.find_element_by_xpath("//a[@class='toggle-content']").click()
        time.sleep(2)
    except NoSuchElementException:
        pass
    
    # scraping the ratings for the hostels
    try:
        rating = driver.find_element_by_xpath("//div[@class='score orange big']")
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append('-')
    
    time.sleep(2)
    
    # scraping the total reviews        
    try:
        rev = driver.find_element_by_xpath("//div[@class='reviews']")
        Reviews.append(rev.text.replace('Total Reviews',''))
    except NoSuchElementException:
        Reviews.append('-')
    
    time.sleep(2)
    
    # scraping the overall review
    try:
        over_rev = driver.find_element_by_xpath("//div[@class='keyword']//span")
        Over_all.append(over_rev.text)
    except NoSuchElementException:
        Over_all.append('-')
    
    time.sleep(2)
    
    # scraping the property description 
    try:
        des = driver.find_element_by_xpath("//div[@class='content']")
        Description.append(des.text)
    except NoSuchElementException:
        Over_all.append('-')
```

```python
#Making dataframe
df = pd.DataFrame({'Hostel Name':Hostel_name,
                'Distance from city centre':Distance,
                'Overall Review':Over_all,
                'Total Reviews':Reviews,
                'Facilities':Facilities,
                'Privates From Price':Pvt_prices,
                'Dorms From Price':Dorms_price,
                'Rating':Rating,
                'Property Description':Description})
df
```
