1. Scrape the details of most viewed videos on YouTube from Wikipedia: Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/ 

You need to find following details:

A) Rank 

B) Name 

C) Artist 

D) Upload date 

E) Views

```python
 #path for chromedriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')

# to open the webpage
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
driver.maximize_window()

time.sleep(2)
```

```python
#empty lists to store the data
name= []
rank= []
artist=[]
upld_dt=[]
views=[]

#scraping the rank
try:
    rnk=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in rnk:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('No details')
    
time.sleep(1)
#scraping the video names
try:
    names=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        name.append(i.text)
except NoSuchElementException:
    name.append('No details')

time.sleep(1)

#scraping the artist name
try:
    art=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in art:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('No details')
    
time.sleep(1)

#scraping the views
try:
    view=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in view:
        views.append(i.text)
except NoSuchElementException:
    views.append('No details')   

time.sleep(1)
#scraping the video upload_date
try:
    date=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in date:
        upld_dt.append(i.text)
except NoSuchElementException:
    upld_dt.append('No details')
    
#Making dataframe
df=pd.DataFrame({"Rank":rank,
                 "Name":name,
                "Artist":artist,
                "Upload date":upld_dt,
                "Views(billions)":views})
df
```

```python
#path for chromedriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')

# to open the webpage
url = 'https://www.bcci.tv/'
driver.get(url)
driver.maximize_window()

time.sleep(2)
```

```python
#to see content of page
soup = BeautifulSoup(page.content)

#clicking the fixtures icon
button=driver.find_element_by_xpath("//div[@class = 'navigation__drop-down drop-down drop-down--reveal-on-hover']/div/ul/li/a")
driver.get(button.get_attribute("href"))
time.sleep(2)

#empty lists to store the data
match_title=[]
series=[]
place=[]
date=[]
dt_time=[]
mtch_time=[]

#scraping the title of the match
for i in driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[@class='u-unskewed-text fixture__format']"):
    try:
        match_title.append(i.text)
    except NoSuchElementException:
        match_title.append('--')

time.sleep(1)
#scraping the series name
for i in driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[@class='u-unskewed-text fixture__tournament-label u-truncated']"):
    try:
        series.append(i.text)
    except NoSuchElementException:
        series.append('--')

time.sleep(1)
#scraping the venue
for i in driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p/span"):
    try:
        place.append(i.text)
    except NoSuchElementException:
        place.append('--')

time.sleep(1)
#scraping the date-time
for i in driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']"):
    try:
        dt_time.append(i.text.replace('\n',' '))
    except NoSuchElementException:
        dt_time.append('--')

#splitting the date and time
date=[i.split(' ',3)[:3] for i in dt_time]
date=[' '.join(i) for i in date]
mtch_time=[i.split(' ',3)[-1] for i in dt_time]

# Creating data frame
df=pd.DataFrame({'Match Title': match_title,
                          "Series": series,
                          "Place": place,
                          "Date": date,
                          "Time": mtch_time})
df
```

5. Scrape the details of trending repositories on Github.com.

Url = https://github.com/

You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

```python
#path for chromedriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')

# to open the webpage
url = 'https://github.com/'
driver.get(url)
driver.maximize_window()

time.sleep(2)
```

```python
# clicking the explore button
exp=driver.find_element_by_xpath("//details[@class='HeaderMenu-details details-overlay details-reset width-full']/summary")
try:
    exp.click()
except:
    driver.get(exp.get_attribute('href'))
time.sleep(1)

#finding the trending option
trend=driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a")
try:
    trend.click()
except:
    driver.get(trend.get_attribute('href'))
time.sleep(1)


#empty lists to store the data   
urls=[]
title=[]
desc=[]
contributor=[]
language=[]


#scraping urls of repositories
repo_url=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in repo_url:
    urls.append(i.get_attribute('href'))

time.sleep(1) 
#scraping title of repositories 
try:
    repo=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
    for i in repo:
        title.append(i.text)
except NoSuchElementException:
    title.append('--')

time.sleep(1)

for i in urls:
    driver.get(i)
    time.sleep(1)
    
    #scraping description of repositories
    try:
        repo_desc = driver.find_element_by_xpath("/html/body/div[4]/div/main/div[2]/div/div/div[2]/div[2]/div/div[1]/div/p")
        desc.append(repo_desc.text)
    except NoSuchElementException:
        desc.append('--')
    
    #scraping the contributors_count
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        contributor.append(count.text)
    except NoSuchElementException:
        contributor.append('--')
    time.sleep(2)
    
     #scraping used languages
    lang=[]
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:    
        for i in languages:
            lang.append(i.text)
    else:
        lang.append('--')
    language.append(lang)
    
time.sleep(1)   
#Making dataframe
df=pd.DataFrame({'Repository Title':title,
                'Repository Description':desc,
                'Contributors Count':contributor,
                'Language Used':language})
df
```

```python
#path for chromedriver
driver = webdriver.Chrome(executable_path='chromedriver.exe')

# to open the webpage
url = 'https://archive.ics.uci.edu/'
driver.get(url)
driver.maximize_window()

time.sleep(2)

#clicking on view all datasets option
view=driver.find_element_by_xpath("//a[@href='datasets.php']")
view.click()

time.sleep(1)
```

```python
#empty lists to store the data    
name=[]
type=[]
task=[]
attribute=[]
no_of_instance=[]
no_of_attribute=[]
year=[]

# Scraping name of the dataset
try:
    names=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in names:
        name.append(i.text)
except NoSuchElementException:
    name.append('--')

time.sleep(1)
# Scraping datatype
try:
    typ=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]")
    for i in typ[1:]:
        type.append(i.text)
except NoSuchElementException:
    type.append('--')
    
time.sleep(1)
# Scraping default task
try:
    tsk=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]")
    for i in tsk[1:]:
        task.append(i.text)
except NoSuchElementException:
    task.append('--')

time.sleep(1)
# Scraping type of attribute
try:
    attri=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]")
    for i in attri[1:]:
        attribute.append(i.text)
except NoSuchElementException:
    attribute.append('--')

time.sleep(1)
#Scraping no. of instances
try:
    inst=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]")
    for i in inst[1:]:
        no_of_instance.append(i.text)
except NoSuchElementException:
    no_of_instance.append('--')
    
time.sleep(1)       
# Scraping no. of attributes
try:
    num=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]")
    for i in num[1:]:
        no_of_attribute.append(i.text)
except NoSuchElementException:
    no_of_attribute.append('--')
        
# Scrapping the year details
try:
    yr=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]")
    for i in yr[1:]:
        year.append(i.text)
except NoSuchElementException:
    year.append('--')
    
# Creating the dataframe
df=pd.DataFrame({"Dataset Name":name,
                 "Data Type":type,
                 "Default Task":task,
                 "Attribute Type":attribute, 
                 "No of Instances":no_of_instance,
                 "No of Attributes":no_of_attribute,
                 "Year":year})
df
```
